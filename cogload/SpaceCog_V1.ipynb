```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np
import seaborn as sns
import matplotlib.image as mpimg 
import plotly.express as px
import plotly.graph_objects as go
from scipy import stats
from plotly.subplots import make_subplots
import pingouin as pg
```

```python
### read raw data ###
file = './UI_VR_Dataset/Sub7_exp_HMD.csv' 
df_Raw= pd.read_csv(file)

df_Raw.head
```

```python
### sample from raw data 
rate = 40
subset = df_Raw[::rate]

UI_On = df_Raw[df_Raw['light_state'] == True].values.tolist()
UI_Off = df_Raw[df_Raw['light_state'] == False].values.tolist()
```

```python
pg.wilcoxon(UI_On, UI_Off, tail='two-sided')
```

```python
## 3D Stastic ###
df = subset


fig = px.scatter_3d(df, x='CameraPosX', y='CameraPosY', z='CameraPosZ',
              color='Trend_Sum',symbol='light_state',opacity=0.8,)
fig.update_traces(marker_size = 3)

fig.show()

# x = df_Raw['CameraPosX']
# y = df_Raw['CameraPosY']
# z = df_Raw['CameraPosZ']
# c = df_Raw['Trend_Sum']
# fig = go.Figure(data=[go.Scatter3d(
#     x=x,
#     y=y,
#     z=z,
#     mode='markers',
#     marker=dict(
#         size=3,
#         # color=c,                # set color to an array/list of desired values
#         colorscale='Viridis',   # choose a colorscale
#         opacity=0.8,
#         # symbol='light_state',
#         marker=dict(size=df_Raw["light_state"],  color=c)
#     )
# )])

# # tight layout
# fig.update_layout(margin=dict(l=0, r=0, b=0, t=0))
# fig.show()
```
