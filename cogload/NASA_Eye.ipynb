```python
### get the start and end periods
df_All = pd.read_csv('D:/code/python/nasa/data/result/Total.csv')
df_All['Strategy'] = df_All['Strategy'].str.strip()
df_All_held = df_All.groupby('Subject')


local_name = locals()
path = 'D:/code/python/nasa/data/csv_cogload/'
# path = 'D:/code/python/nasa/data/test/'
colors = ['B','W','P','G']
axis3 = ['X','Y','Z']
groups = ['control','Anchoring','Asynchronized','Synchronized']

column_W = 'Cube6'
column_B = 'Cube5'
column_G = 'Cube4'
column_P = 'Cube1'
object_W = 'Object_White'
object_B = 'Object4_Blue'
object_G = 'Object4_Green'
object_P = 'Object_Purple'

period_length = 900

for color in colors:
    local_name[color + '_df_start'] = pd.DataFrame()
    local_name[color + '_df_end'] = pd.DataFrame()
    for group in groups:
        local_name[color + '_' + group + '_df_start'] = pd.DataFrame()
        local_name[color + '_' + group + '_df_end'] = pd.DataFrame()

for home, dirs, files in os.walk(path):
    for filename in files:
        subject = filename[1:3]
        trail = filename[5:-8]
        if subject != '28' and trail != 'raining' and trail != 'raining ':
            df = pd.read_csv(home+"/"+filename,index_col=0)
            df_groupby_held = df.groupby('CollisionTargetName')
            labels = df_groupby_held.size().index

            df_c = df_All_held.get_group(int(subject))
            i = int(trail) - 1
            id = str(subject)+ df_c['Strategy'].values[i]+str(i)
            group = df_c['Strategy'].values[i]

            for color in colors:
                if local_name['object_'+ color] not in labels:
                    local_name[color + '_' + group + '_df_start'] = local_name[color + '_' + group + '_df_start'].append(pd.DataFrame())
                    local_name[color + '_' + group + '_df_end'] = local_name[color + '_' + group + '_df_end'].append(pd.DataFrame())
                else:
                    local_name[color + '_' + group + 'start_point'] = df.CollisionTargetName.isin([local_name['object_'+color]]).idxmax()
                    local_name[color + '_' + group + 'end_point'] = df[::-1].CollisionTargetName.isin([local_name['object_'+color]]).idxmax()
                    
                    if local_name[color + '_' + group + 'start_point'] - period_length < 0:
                        local_name[color + '_' + group + 'start_period'] = df[0 : local_name[color + '_' + group + 'start_point']]
                    else:
                        local_name[color + '_' + group + 'start_period'] = df[local_name[color + '_' + group + 'start_point'] - period_length : local_name[color + '_' + group + 'start_point']]
                    local_name[color + '_' + group + 'start_period']['data_state'] = 'before'
                    if local_name[color + '_' + group + 'start_point'] + period_length > len(df):
                        local_name[color + '_' + group + 'start_period_after'] = df[local_name[color + '_' + group + 'start_point'] : len(df)]
                    else:
                        local_name[color + '_' + group + 'start_period_after'] = df[local_name[color + '_' + group + 'start_point'] : local_name[color + '_' + group + 'start_point'] + period_length]
                    local_name[color + '_' + group + 'start_period_after']['data_state'] = 'after'
                    local_name[color + '_' + group + 'start_period'] = local_name[color + '_' + group + 'start_period'].append(local_name[color + '_' + group + 'start_period_after'])
                    
                    local_name[color + '_' + group + 'end_period'] = df[local_name[color + '_' + group + 'end_point'] - period_length : local_name[color + '_' + group + 'end_point']]
                    local_name[color + '_' + group + 'end_period']['data_state'] = 'before'
                    local_name[color + '_' + group + 'end_period_after'] = df[local_name[color + '_' + group + 'end_point'] : local_name[color + '_' + group + 'end_point'] + period_length]
                    local_name[color + '_' + group + 'end_period_after']['data_state'] = 'after'
                    local_name[color + '_' + group + 'end_period'] = local_name[color + '_' + group + 'end_period'].append(local_name[color + '_' + group + 'end_period_after'])

                    if local_name[color + '_' + group + 'end_point'] - local_name[color + '_' + group + 'start_point'] > period_length:
                        length_state = 'long_enough'
                    else:
                        length_state = 'not_long_enough'

                    local_name[color + '_' + group + 'start_period']['ID'] = id
                    local_name[color + '_' + group + 'start_period']['Conditions'] = df_c['Strategy'].values[i]
                    local_name[color + '_' + group + 'start_period']['SubjectNO'] = str(subject)
                    local_name[color + '_' + group + 'start_period']['Point_state'] = 'Start'
                    local_name[color + '_' + group + 'start_period']['Length_state'] = length_state

                    local_name[color + '_' + group + 'end_period']['ID'] = id
                    local_name[color + '_' + group + 'end_period']['Conditions'] = df_c['Strategy'].values[i]
                    local_name[color + '_' + group + 'end_period']['SubjectNO'] = str(subject)
                    local_name[color + '_' + group + 'end_period']['Point_state'] = 'End'
                    local_name[color + '_' + group + 'end_period']['Length_state'] = length_state

                    local_name[color + '_' + group + '_df_start'] = local_name[color + '_' + group + '_df_start'].append(local_name[color + '_' + group + 'start_period'])
                    local_name[color + '_' + group + '_df_end'] = local_name[color + '_' + group + '_df_end'].append(local_name[color + '_' + group + 'end_period'])

                    
for color in colors:
    for group in groups:
        local_name[color + '_' + group + '_df_start'].to_csv(path + color + '_' + group + '_df_start.csv')
        local_name[color + '_' + group + '_df_end'].to_csv(path + color + '_' + group + '_df_end.csv')

        local_name[color + '_df_start'] = local_name[color + '_df_start'].append(local_name[color + '_' + group + '_df_start'])
        local_name[color + '_df_end'] = local_name[color + '_df_end'].append(local_name[color + '_' + group + '_df_end'])
    local_name[color + '_df_start'].to_csv(path + color + '_df_start.csv')
    local_name[color + '_df_end'].to_csv(path + color + '_df_end.csv')

```
